```python
#to download  motor bike images using bing-image -downloader with a limit of 150 images
from bing_image_downloader import downloader
downloader.download("motorbikeimages", limit= 150, output_dir = 'bikimages')
```

**importing libraries ** computer vision, matplot, draw_bbox.

```python
import cv2 
!pip install cvlib
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
```

```python
#read the image using imread fuction in cv2 and plt the image using imshow function
im= cv2.imread("/content/bikimages/motorbikeimages/Image_100.JPG")
plt.imshow(im)
```

```python
#to convert the image in actual colur by using open cv
image= cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#detect_common _objects method is used to detect common object in the image 
bbox, label, conf = cv.detect_common_objects(im, confidence=0.6, nms_thresh=0.3, model='yolov3')
```

```python
#draw_bbox method is used to apply bounding box to the image
output_image= draw_bbox(image,bbox, label, confidence=0.6)
```

```python
#display the image 
plt.imshow(output_image)
plt.show()
```
